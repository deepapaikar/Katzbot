```python
from pprint import pprint
import nltk
import os
from tqdm import tqdm
import re
from Questgen import main
qe= main.BoolQGen ()
```

```python
def generate_bool_QA(text):
    payload = {
            "input_text": text
        }
    output = qe.predict_boolq(payload)
    return output['Boolean Questions']

def generate_answer(text, *questions):
    predictor = main.AnswerPredictor(model_name="T5")
    answers = []
    for ques in questions:
        payload = {
            "input_text": text,
            "input_question": ques
        }   
        answers.append(predictor.predict_answer(payload))
    return answers


Summer_data = ((i, open(f"Summer text data/"+i, encoding='utf-8').readlines()) for i in os.listdir("Summer text data"))
```

```python
text = re.sub("\n", " ", " ".join(next(Summer_data)[1]))
text
```

```python
questions = generate_bool_QA(text)
pprint(["Questions:", questions])
pprint(generate_answer(text, *questions))
```

```python
master_questions = []
master_answers = []
for title, text in tqdm(Summer_data, total = len(os.listdir("Summer text data"))):
    text = re.sub("\n | [^A-Za-z0-9\s]", " ", "".join(text))
    questions = generate_bool_QA(text)
    master_questions.extend(questions)

    answers = generate_answer(text, *questions)
    master_answers.extend(answers)
```
